```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.array([2023,2024,2025,2026])
y=np.array([15,25,30,20])

# plt.plot(x,y)
plt.plot(y)

plt.show()
```

```python
x=[2023,2024,2025,2026]
y=[15,25,30,20]

plt.plot(x,y , marker=".",
                markersize=20,
                markerfacecolor="black",
                markeredgecolor="black",
                linestyle="dotted",#dashes/dashdot/solid/none
                linewidth=4,
                color="blue")

plt.show()
```

```python
x=[2023,2024,2025,2026]
y1=[15,25,30,20]
y2=[17,23,35,10]

#make a directory for all styles 
line_style=dict(marker=".",
                markersize=20,
                markerfacecolor="black",
                markeredgecolor="black",
                # linestyle="solid",#dashes/dashdot/solid/none
                linewidth=2,
                color="blue")

plt.plot(x,y1 ,**line_style) #call the line_style directory 

plt.plot(x,y2, **line_style ,linestyle="dotted")

plt.show()
```

LABLES IN PYTHON

```python
x=[2023,2024,2025,2026]
y1=[15,25,30,20]
y2=[17,23,35,10]

plt.title("class size",fontsize=20 ,family="Arial",fontweight="bold",color="red")

plt.xlabel("Year",fontsize=20 ,family="Arial",fontweight="bold",color="blue")

plt.ylabel("NO of studests",fontsize=10 ,family="Arial",fontweight="bold",color="black")

plt.tick_params(axis="both",color='green')

plt.plot(x,y1)
plt.plot(x,y2)

plt.xticks(x)
plt.show()
```

GRID LINE

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]

plt.grid(axis="both",
         linewidth=2,
         color="gray",
         linestyle="dotted")
plt.plot(x,y)
plt.show()
```

BAR Chart

```python
cate=np.array(["grains","fruits","vegies","protein","dairy","sweet"])

values=np.array([4,3,5,6,7,2])

# plt.bar(cate,values,color="skyblue")
plt.barh(cate,values,color="skyblue")

plt.title("Daily consumption")

plt.xlabel("categories")

plt.ylabel("values")

plt.show()
```

pie chart

```python
cate=np.array(["grains","fruits","vegies","protein"])

values=np.array([40,30,50,60])
colour=np.array(["red","black","blue","green"])
plt.pie(values,labels=cate,
                            autopct="%1.1f%%",
                            colors=colour,
                            explode=[0,0,0,0.1],
                            shadow=True,
                            startangle=90)

plt.title("Daily consumption")
```

SCATTER Graph

```python
x=[0,1,1,2,3,4,5,6,7,7,8]

y=[55,60,65,62,68,70,75,78,82,85,97]

x2=[0,1,1,2,3,4,5,6,7,7,8]

y2=[50,60,58,55,60,70,73,70,80,80,90]

plt.scatter(x,y,color="black",alpha=0.2,s=200,label="class a")
plt.scatter(x2,y2,color="red",alpha=0.2,s=200,label="class b")

plt.xlabel("hours of study")
plt.ylabel("graid")
plt.title("test score")
plt.legend()
plt.show()
```

HISTOGRAM

```python
scores=np.random.normal(loc=80,scale=10,size=100)

plt.hist(scores,bins=10,color="green",edgecolor="black")


plt.show()
```

SUBPLOT in python

```python
figure,axes=plt.subplots(2,2)

axes[0,0].plot(x,y,color="red")
axes[0,0].set_title("X")

axes[0,1].bar(x,y,color="green")
axes[0,1].set_title("0,1")

axes[1,0].scatter(x,y,color="blue")
axes[1,0].set_title("1,0")

axes[1,1].hist(scores,color="purple")
axes[1,1].set_title("1,1")
plt.show()
```
