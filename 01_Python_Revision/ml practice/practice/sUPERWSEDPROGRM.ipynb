SUPERWISED LEARNING


LINEAR Regression

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
X=[[2],[4],[6],[8]]
y=[65,70,85,90]
model.fit(X,y)

value=float(input("Entervalue to predict"))
prediction=model.predict([[value]])

print(f"you studied {value},and you will get around {prediction} marks")

```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression()
X=[[2],[4],[6],[8]]
y=[0,0,1,1]
model.fit(X,y)

value=float(input("Entervalue to predict"))
prediction=model.predict([[value]])[0]

if prediction==1:
    print(f"you are likely to pass")
else:
    print("you will probabily fail")
```

KNN(K_nearest_Naighbors) lazy learner

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

X=[
    [7,2],
    [8,3],
    [9,8],
    [10,9]
]
y=[0,0,1,1]

model= DecisionTreeClassifier()

model.fit(X,y)

size=input("enmter size")
shade=input("enter shade")
predection = model.predict([[size,shade]])[0]

if predection==0:
    print("this is apple")
else:
    print("This is orange")
```

Classifiction Matrics

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

y_true=[1,0,1,0,1,1]
y_pred=[1,1,1,0,1,0]

print("Accuracy Score ",accuracy_score(y_true,y_pred))
print("Pression Score ",precision_score(y_true,y_pred))
print("Recall Score ",recall_score(y_true,y_pred))
print("f1 Score ",f1_score(y_true,y_pred))
```

Confusion Matrics


```python
from sklearn.metrics  import confusion_matrix
y_true=[1,0,1,1,0,1,1,0,1]
y_pred=[1,0,1,0,1,0,1,0,0]

cm=confusion_matrix(y_true,y_pred)

print('Confusion Matrix',cm)
```

Regression Matrics (MAE MSE RMSE)

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
import numpy as np

real_score=[90,60,80,100]
Predected_score=[85,70,70,95]

mae=mean_absolute_error(real_score,Predected_score)
mse=mean_squared_error(real_score,Predected_score)

RMSC=np.sqrt(mse)

print(mae)
print(mse)
print(RMSC)
```
