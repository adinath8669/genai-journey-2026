Handling missing data

```python
df =pd.read_csv("orders.csv")

df.dropna()
df.fillna("NULL")
df.isnull().sum()

df.isnull().mean() * 100
```

encoding categorical value

1.label encoding
2.onehot encoding

```python
#lable encoding
from sklearn.preprocessing import LabelEncoder
df_label=df.copy()

le=LabelEncoder()

df_label["shipped_data"]=le.fit_transform(df_label["Shipped"])

print(df_label[['CustomerName','Quantity','Price','Shipped','shipped_data']])
```

```python
#One hot encoding

df_encoded=pd.get_dummies(df_label,columns=['Country'])

print(df_encoded)
```

Splitting Data

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split

data={
    'studyHours':[1,2,3,4,5],
    "testScore":[40,50,70,90,95]
}

df=pd.DataFrame(data)

standard_scale=StandardScaler()
standard=standard_scale.fit_transform(df)

print(pd.DataFrame(standard, columns=['studyHours','testScore']))

print("*"*50)

min_max=MinMaxScaler()
min_max_Scaled=min_max.fit_transform(df)

print(pd.DataFrame(min_max_Scaled, columns=['studyHours','testScore']))


print("*"*50)

X=df[['studyHours']]
y=df[['testScore']]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

print(X_train)
print(X_test)
```
